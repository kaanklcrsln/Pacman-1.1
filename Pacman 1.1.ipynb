```python
import pygame

pygame.init()

screen_width = 800
screen_height = 800
screen = pygame.display.set_mode((screen_width, screen_height))

pygame.display.set_caption("Pac-Man")
icon = pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png')
pygame.display.set_icon(icon)

player_image = pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png')
dot_image = pygame.image.load(r'C:\Users\ASUS\Desktop\dot.png')

player_x = 300
player_y = 300
dot_x = 100
dot_y = 100

running = True
score = 0

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                player_x -= 5
            elif event.key == pygame.K_RIGHT:
                player_x += 5
            elif event.key == pygame.K_UP:
                player_y -= 5
            elif event.key == pygame.K_DOWN:
                player_y += 5
    if abs(player_x - dot_x) < 20 and abs(player_y - dot_y) < 20:
        score += 1
        dot_x = -100
        dot_y = -100
    
    screen.fill((0, 0, 0))
    screen.blit(player_image, (player_x, player_y))
    screen.blit(dot_image, (dot_x, dot_y))
    font = pygame.font.Font("freesansbold.ttf", 32)
    text = font.render("Score: " + str(score), True, (255, 255, 255))
    screen.blit(text, (10, 10))
    pygame.display.flip()

pygame.quit()
```

```python
import pygame
import random

pygame.init()

screen_width = 800
screen_height = 800
screen = pygame.display.set_mode((screen_width, screen_height))

pygame.display.set_caption("Pac-Man")
icon = pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png')
pygame.display.set_icon(icon)

player_image = pygame.transform.scale(pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png'), (40, 40))
dot_image = pygame.transform.scale(pygame.image.load(r'C:\Users\ASUS\Desktop\dot.png'), (20, 20))

player_x = 300
player_y = 300

dots = []

player_speed = 3
player_x_change = 0
player_y_change = 0

running = True
score = 0

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                player_x_change = -player_speed
            elif event.key == pygame.K_RIGHT:
                player_x_change = player_speed
            elif event.key == pygame.K_UP:
                player_y_change = -player_speed
            elif event.key == pygame.K_DOWN:
                player_y_change = player_speed
        elif event.type == pygame.KEYUP:
            
            if event.key in (pygame.K_LEFT, pygame.K_RIGHT):
                player_x_change = 0
            if event.key in (pygame.K_UP, pygame.K_DOWN):
                player_y_change = 0

    player_x += player_x_change
    player_y += player_y_change

    if len(dots) < 5: 
        for _ in range(5 - len(dots)):
            dots.append({'x': random.randint(0, screen_width - 20), 'y': random.randint(0, screen_height - 20)})
    for dot in dots:
        if player_x < dot['x'] + 20 and player_x + 40 > dot['x'] and player_y < dot['y'] + 20 and player_y + 40 > dot['y']:
            score += 1
            dots.remove(dot)

    screen.fill((0, 0, 0))

    screen.blit(player_image, (player_x, player_y))

    for dot in dots:
        screen.blit(dot_image, (dot['x'], dot['y']))

    font = pygame.font.Font(None, 32)
    text = font.render("Score: " + str(score), True, (255, 255, 255))
    screen.blit(text, (10, 10))
    pygame.display.flip()

pygame.quit()
```

```python
import pygame
import random

pygame.init()

screen_width = 1280
screen_height = 720
screen = pygame.display.set_mode((screen_width, screen_height))

pygame.display.set_caption("Pac-Man")
icon = pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png')
pygame.display.set_icon(icon)

player_image = pygame.transform.scale(pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png'), (40, 40))
dot_image = pygame.transform.scale(pygame.image.load(r'C:\Users\ASUS\Desktop\dot.png'), (20, 20))

player_x, player_y = 300, 300
player_speed = 2
dots, score = [], 0

running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    player_x += (keys[pygame.K_RIGHT] - keys[pygame.K_LEFT]) * player_speed
    player_y += (keys[pygame.K_DOWN] - keys[pygame.K_UP]) * player_speed

    if len(dots) < 5:
        dots.append({'x': random.randint(0, screen_width - 20), 'y': random.randint(0, screen_height - 20)})

    for dot in dots:
        if pygame.Rect(player_x, player_y, 40, 40).colliderect(pygame.Rect(dot['x'], dot['y'], 20, 20)):
            score += 1
            dots.remove(dot)

    screen.fill((0, 0, 0))
    screen.blit(player_image, (player_x, player_y))

    for dot in dots:
        screen.blit(dot_image, (dot['x'], dot['y']))

    font = pygame.font.Font(None, 32)
    text = font.render(f"Score: {score}", True, (255, 255, 255))
    screen.blit(text, (10, 10))
    pygame.display.flip()

pygame.quit()
```

```python
import pygame
import random

pygame.init()

screen_width = 800
screen_height = 800
screen = pygame.display.set_mode((screen_width, screen_height))

pygame.display.set_caption("Pac-Man")
icon = pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png')
pygame.display.set_icon(icon)

dot_image = pygame.transform.scale(pygame.image.load(r'C:\Users\ASUS\Desktop\dot.png'), (20, 20))
dots = []

player_size = 40
player_image = pygame.transform.scale(pygame.image.load(r'C:\Users\ASUS\Desktop\pac-man-512x512.png'), (player_size, player_size))
player_x, player_y = 300, 300
player_speed = 3

running = True
score = 0

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    player_x += (keys[pygame.K_RIGHT] - keys[pygame.K_LEFT]) * player_speed
    player_y += (keys[pygame.K_DOWN] - keys[pygame.K_UP]) * player_speed

    if len(dots) < 5:
        dots.append({'x': random.randint(0, screen_width - 20), 'y': random.randint(0, screen_height - 20)})

    for dot in dots:
        if pygame.Rect(player_x, player_y, player_size, player_size).colliderect(pygame.Rect(dot['x'], dot['y'], 20, 20)):
            score += 1
            dots.remove(dot)
            player_size += 2  # Her puan toplandığında Pac-Man boyutunu büyütün
            player_image = pygame.transform.scale(player_image, (player_size, player_size))

    screen.fill((0, 0, 0))
    screen.blit(player_image, (player_x, player_y))

    for dot in dots:
        screen.blit(dot_image, (dot['x'], dot['y']))

    font = pygame.font.Font(None, 32)
    text = font.render(f"Score: {score}", True, (255, 255, 255))
    screen.blit(text, (10, 10))
    pygame.display.flip()

pygame.quit()
```
